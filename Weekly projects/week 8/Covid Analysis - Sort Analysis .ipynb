### * Which Country is having maximum confirmed cases

# method 1 

```python
# maximum confirmed cases in a country 

for i in df.sort_values(by = 'Confirmed',ascending = False).head(1).values:
    
    print('Country having maximum confirmed cases : ',i[0])
    print('Confirmed cases                        : ',i[1])
```

```python
# top country sorted by the confirmed cases due to COVID-19 
df.sort_values(by = 'Confirmed' ,ascending = False).head()
```

# method 2 :- Using arrays 

```python
data = df.values
lst = []
for i in data:
    lst.append(i[1])
for i in data:
    if (i[1] == max(lst)):
        print('Country having maximum confirmed cases : ',i[0])
        print('Confirmed cases                        : ',i[1])
```


### * Which Country is having maximum deaths?

```python
for i in df.sort_values(by = 'Deaths',ascending = False).head(1).values:
    print('Country with maximum death          : ', i[0])
    print('Death cases                         : ', i[2])
```

```python
# top country sorted by the death rates due to COVID-19 
df.sort_values(by = 'Deaths',ascending = False).head() # shows top 5 countries 
```
