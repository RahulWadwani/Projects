# Add Month Column in the dataset from Date Column (Jan-Dec)

```python
dct = {1:'January',
       2:'February',
       3:'March',
       4: 'April',
       5:'May',
       6:'June',
       7:'July',
       8:'August',
       9:'Septemeber',
       10:'October',
       11:'November',
       12:'December'}
```

```python
month = []
for i in df['date'].values:
    month.append(i.split('-')[1])
    
new_month = []
for i in month:
    for j in dct.keys():
        if (int(i) == j):
            new_month.append(dct[j])
            

df['Month'] = new_month
df.head()
```
