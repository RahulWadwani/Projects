# Importing and reading the file 

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('product_sale.csv')
df
```

### 1) Product and total product sold | pie chart and bar graph

```python
fig,axs = plt.subplots(figsize = (10,8))
plt.pie(df['total_prod_sold_to'],labels = df['product'],autopct = '%1.2f%%')
plt.show()
```

```python
fig,axs = plt.subplots(figsize = (10,8))
plt.barh(df['product'],df['total_prod_sold_to'])
plt.xlabel('Total Product Sold')
plt.ylabel('Products')
plt.show()
```

### 2) Product and total quantity sold | pie chart and bar graph

```python
fig,axs = plt.subplots(figsize = (10,8))
plt.pie(df['total_quantity_sold'],labels = df['product'],autopct = '%1.2f%%')
plt.show()
```

```python
fig,axs = plt.subplots(figsize = (10,8))
plt.barh(df['product'],df['total_quantity_sold'])
plt.xlabel('Total quantity Sold')
plt.ylabel('Products')
plt.show()
```

### 3) Product and total_amount | pie chart and bar graph

```python
fig,axs = plt.subplots(figsize = (10,8))
plt.pie(df['total_amount'],labels = df['product'],autopct = '%1.2f%%')
plt.show()
```
