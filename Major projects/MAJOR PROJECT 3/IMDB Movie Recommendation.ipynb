# Importing dataset 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Data-Science-Specialization/main/Week%20-12/IMDB%20-%20Most%20Popular%20Movie%20%26%20Series/final.csv')
df.head()
```

```python
## preprocessing the genre column
data = []

for i in df.values:
    if (i[3] != 'Nan'):
        data.append([i[0],i[1],i[2],i[3],i[4],i[5],i[6]])
        
df = pd.DataFrame(data,columns = ['name','year','duration','genre','rating','director','type'])
df
```

```python
genres = []
for i in df['genre']:
    genres += (i.split(','))
genres = list(set(genres))
```

```python
one_hot = []
for i in df['genre']:
    i = [j.strip() for j in i.split(',')]
    enc = [0 for encode in range(len(genres))]
    for genre in i:
        enc[genres.index(genre)] = 1
    one_hot.append(enc)
```

```python
one_hot = np.array(one_hot)
```

```python
for i in range(len(genres)):
    df[genres[i]] = one_hot[:,i]
    

df.head()
```

```python
for i in range(len(genres)):
    df[genres[i]] = one_hot[:,i]
    
df.head()
```

## 4. Also recommend the movies based on the name of the movie (you need to perform one hot encoding on name column of the dataset and finding the similarity) for example The Godfather and The Godfather - II

```python
movies = []
for i in df['name']:
    movies += i.split(' ')
movies = (list(set(movies)))
```

```python
hot_encode = []

for i in df['name']:
    i = ([words.strip() for words in i.split(' ')])
    enco = [0] * len(movies)
    for movie in i:
        enco[movies.index(movie)] = 1
    hot_encode.append(enco)
    
hot_encode = np.array(hot_encode)
```

```python
for i in range(len(movies)):
    df[movies[i]] = hot_encode[:,i]
    
df.head()
```

```python
def mod(lst):
    s = 0
    for i in lst:
        if i < 0:
            s += -i 
        else:
            s += i
    return s
```
