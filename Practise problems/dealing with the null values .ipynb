# 3.Imputer 

```python
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('data.csv',header = None)
```

```python
df
```

```python
English = np.reshape(df[2].values,(11,1))
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(English)
df['English'] = imputer.transform(English)


Hindi = np.reshape(df[1].values,(11,1))
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(Hindi)
df['Hindi'] = imputer.transform(Hindi)



maths = np.reshape(df[3].values,(11,1))
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(maths)
df['maths'] = imputer.transform(maths)


science = np.reshape(df[4].values,(11,1))
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(science)
df['science'] = imputer.transform(science)


ss = np.reshape(df[5].values,(11,1))
imputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imputer.fit(ss)
df['ss'] = imputer.transform(ss)
```
